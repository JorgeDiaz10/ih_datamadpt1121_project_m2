```python
query = '''SELECT * FROM diamonds_dimensions'''
query2 = '''SELECT * FROM diamonds_transactional'''
query3 = '''SELECT * FROM diamonds_properties'''
query4 = '''SELECT * FROM diamonds_cut'''
query5 = '''SELECT * FROM diamonds_color'''
query6 = '''SELECT * FROM diamonds_clarity'''
query7 = '''SELECT * FROM diamonds_city'''
```

```python
df = pd.read_sql_query(query, connection)
```

```python
df2 = pd.read_sql_query(query3, connection)
```

```python
df3 = pd.read_sql_query(query4, connection)
```

```python
df4 = pd.read_sql_query(query5, connection)
```

```python
df5 = pd.read_sql_query(query6, connection)
```

```python
df6 = pd.read_sql_query(query7, connection)
```

```python
df_unido1 = pd.merge(df, df1, on = 'index_id')
df_unido2 = pd.merge(df_unido1,df2, on = 'index_id')
df_unido3 = pd.merge(df_unido2,df3, on = 'cut_id')
df_unido4 = pd.merge(df_unido3,df4, on = 'color_id')
df_unido5 = pd.merge(df_unido4,df5, on = 'clarity_id')
df_unido6 = pd.merge(df_unido5,df6, on = 'city_id')
df_unido6
```

```python
df_final = df_unido6.loc[:, ['index_id', 'depth', 'table', 'x', 'y', 'z', 'price', 'carat', 'cut', 'color', 'clarity', 'city']]
df_final
```

```python
query_sql = '''SELECT trans.index_id, cla.clarity, col.color, ct.cut, diam.depth, 
diam.'table', diam.x, diam.y, diam.z, trans.price, ci.city, trans.carat
FROM diamonds_properties pro
   INNER JOIN diamonds_clarity cla ON pro.clarity_id = cla.clarity_id 
   INNER JOIN diamonds_color col on pro.color_id = col.color_id 
   INNER JOIN diamonds_cut ct on pro.cut_id = ct.cut_id 
   INNER JOIN diamonds_dimensions diam on pro.index_id = diam.index_id
   INNER JOIN diamonds_transactional trans on pro.index_id = trans.index_id
   INNER JOIN diamonds_city ci on trans.city_id = ci.city_id'''
```

```python
df_sql = pd.read_sql_query(query_sql, connection)
df_sql.head()
```

```python
df_sql.rename(columns = {'x': 'length', 'y': 'width', 'z': 'height'}, inplace = True)
df_sql.head()
```
